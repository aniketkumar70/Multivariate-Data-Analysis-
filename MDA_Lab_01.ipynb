### Aim: To learn the basics of libraries used such as pandas, matplotlib, numpy, etc.

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
print("Matplotlib:",mpl.__version__)
print("Numpy:",np.__version__)
print("Pandas:", pd.__version__)
```

```python
data=pd.read_csv("canadian_immegration_data.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
data.describe().transpose()
```

```python
print("Continent:",data['Continent'].unique())
print("Regions:",data['Region'].unique())
```

```python
data.set_index('Country',inplace=True)
data.index.name=None
data.columns
```

```python
data
```

```python
data
```

```python
data.sort_values(by='Total', inplace=True, ascending=True)
top_ten=data['Total'].tail(10)
top_ten
```

```python
top_ten.plot(kind="barh",figsize=(10,6),color='steelblue')
plt.xlabel("Number of Immigrants")
plt.title("Top 10 countries of immigration  to Canada in between 1980-2013")
```

```python
continents_df=data.groupby('Continent').sum()
continents_df.head()

continents_df['Total'].plot(kind="pie",figsize=(6,6))
plt.title("Immigration to Canada by Continent from 1980-2013")
```
