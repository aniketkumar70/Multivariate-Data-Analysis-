```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_excel("HeightWeight.xlsx")
data
```

```python
height=data.Height
weight=data.Weight
```

## Scatterplot

```python
plt.scatter(height,weight)
plt.xlabel("Height")
plt.ylabel("Weight")
plt.title("Height vs. Weight")
plt.show()
```

```python
[m,c]=np.polyfit(height, weight, deg=1)
plt.plot(height, m*height+c)
plt.scatter(height, weight)
plt.show()
```

#### The linear model approximately fits our data.

## Finding mean vector

```python
X=np.array([height,weight]).transpose()
n=len(X)
ones=np.ones(n)
meanVec=1/n*X.transpose().dot(ones)
meanVec
```
